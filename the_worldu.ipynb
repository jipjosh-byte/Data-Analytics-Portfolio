```python
col = 'IMF_Estimate'
Q1 = gdp_df[col].quantile(0.25)
Q3 = gdp_df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = gdp_df[(gdp_df[col] < lower) | (gdp_df[col] > upper)]
```

```python
import matplotlib.pyplot as plt
numeric_cols = ['IMF_Estimate', 'WorldBank_Estimate', 'UN_Estimate']
```

```python
for col in numeric_cols:
  plt.figure(figsize=(8, 4))
  plt.boxplot(gdp_df[col].dropna(), vert=False)  # dropna avoids errors from missing values
  plt.title(f'Boxplot of {col} (outliers shown as dots)')
  plt.xlabel(col)
  plt.show()
```

```python
print(f"Number of outliers: {len(outliers)}")

print(outliers[['Country/Territory', col]].sort_values(col, ascending=False))
```

### Advanced Analysis: Distribution Density
Let's use a Violin Plot to compare the distribution of marks across different classes, separated by gender. This helps identify if certain classes or genders have more consistent or varied performance.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='class', y='mark', hue='gender', data=stu_df, split=True, palette='Pastel1')
plt.title('Distribution of Marks by Class and Gender')
plt.xlabel('Class')
plt.ylabel('Marks')
plt.legend(title='Gender')
plt.show()
```

### 1. Scatter Plot
This is a scatter plot of ID versus Mark, and this is what it shows: how individual student scores are distributed across the dataset, helping to spot any sequential trends or outliers.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='id', y='mark', data=stu_df, hue='class', palette='viridis')
plt.title('Student ID vs Marks')
plt.show()
```

### 2. Stacked Bar Chart
This is a bar chart showing the student count by class, and this is what it shows: the distribution of student enrollment across different grade levels.

```python
plt.figure(figsize=(10, 6))
stu_df['class'].value_counts().plot(kind='bar', color='teal', edgecolor='black')
plt.title('Number of Students per Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

### 3. Histogram
This is a histogram for marks, and this is what it shows: the frequency of different score ranges, allowing you to see how many students fall into specific performance brackets.

```python
plt.figure(figsize=(10, 6))
plt.hist(stu_df['mark'], bins=10, color='red',)
plt.title('Histogram of Student Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

### 4. Pie Chart
This is a pie chart of gender distribution, and this is what it shows: the proportion of male versus female students in the dataset.

```python
plt.figure(figsize=(8, 8))
stu_df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Gender Ratio')
plt.ylabel('')
plt.show()
```

### 5. Heatmap
This is a heatmap of average marks by Class and Gender, and this is what it shows: which specific groups (e.g., Females in Class Four) are performing the best at a glance.

```python
pivot_table = stu_df.pivot_table(values='mark', index='class', columns='gender', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='Reds')
plt.title('Heatmap of Average Marks by Class and Gender')
plt.show()
```
